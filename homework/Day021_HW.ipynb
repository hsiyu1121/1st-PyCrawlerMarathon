```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import datetime
```

## 根據範例，取出今天所有的新聞

```python
today=time.strftime('%Y/%m/%d',time.gmtime())

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(80):
    time.sleep(5)
    browser.execute_script("window.scrollTo(0, 30000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find(class_='date').text)[:10] == today:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        break
        
# 關閉瀏覽器
browser.quit();
```

## 取出今天下午三點到五點的新聞

```python

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

tags_h3=soup.find(class_="part_list_2").find_all('h3')

for i in range(80):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 30000);")
    

for tag_h3 in tags_h3:
    hour = int((tag_h3.find(class_='date').text)[11:13]) 
    minu = int((tag_h3.find(class_='date').text)[14:16]) 

    if (hour == 15 or hour == 16) or (hour==17 and minu==0) and (tag_h3.find(class_='date').text)[:10] == today:
        print(tag_h3.find(class_="date").text, tag_h3.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```
