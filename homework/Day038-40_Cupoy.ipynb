```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
```

```python
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get("https://www.cupoy.com/newsfeed/topstory")


for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
count=1
total=[]
data_list=[]


while count <= 500:
    html_source = browser.page_source
    soup=BeautifulSoup(html_source,'html5lib')
    #hot_News=soup.find('div', class_='ReactVirtualized__Grid__innerScrollContainer').find_all(class_='sc-eEieub sc-iuDHTM ibJqYc')
    hot_News=soup.find_all(class_='sc-eEieub sc-iuDHTM ibJqYc')
    for item in hot_News:
        category = item.find(class_ ="sc-gacfCG bPSpUf").text
        title = item.find(class_ ="sc-erNlkL sc-ekulBa hDLssh").text
        content = item.find('p').text
        
        if item not in total:
            print(count, category, title)
            count+=1
            
            total.append({
                '類別': category,
                '標題': title,
                '內容': content
                })

            data_list.append(category)
        if count == 501:
            break
            
# 關閉瀏覽器
browser.quit();
```

```python
df=pd.DataFrame(total)
df
```

```python
df2 = pd.DataFrame(df.groupby('類別').size(), columns=['數量'])
df2
```
